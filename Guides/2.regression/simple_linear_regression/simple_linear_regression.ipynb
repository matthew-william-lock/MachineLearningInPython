# Simple Linear Regression
Simple linear regression takes the form of fitting to a trendline, specifically in this case a linear trendline. 
<br>
Therefore this model applies to data with linear correlation.
 

$y = b_0+b_1*x_1$ 

- $y$ is the Dependent Variable (DV)
- $x_1$ is the Independent Variable (IV)
- The independent variable may directly or indirectly affect the DP
- $b_1$ is the Coefficient (coefficient can be considered to be coefficient of proportion) 
- $b_0$ is the constant 

This guide deals with predicting Salary by the ammount of Experience using Simple Linear Regression. <br> <br>
<img src="images/simple_linear_reg.png" width="300"> 
<br>
<br>
where our equation becomes $y = b_0+b_1*Experience$ 
- $b_0$ becomes the intersection of the y axis
- $b_1$ becomes the gradient of the line

- where $y_i$ is an actual observation, and $\hat{y_i}$ is the model observation. 
- the green line idicates the difference between the actual and model observations

Modeling essentially invlolves minimizing all of these distances: 


minimize $\sum_{i=0}^{N} (y_i-\hat{y_i})^2$

To do these we use the **Ordinary Least Squares Method**



## Importing the libraries

```
# General libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Split the dataset
from sklearn.model_selection import train_test_split

# Simple Linear Model
from sklearn.linear_model import LinearRegression
```

## Importing the dataset

```
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```
regressor = LinearRegression() # No paramters are required
regressor.fit(X_train,y_train) # This will train the model to the training set

# The output produced by this block confirms the successful creation of the regression model
# Your linear regression model has now been trained!
```

## Predicting the Test set results

```
y_pred = regressor.predict(X_test) # Create vector of predicted salaries
```

## Visualising the Training set results

```
plt.scatter(X_train, y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (R)')
plt.show()
```

## Visualising the Test set results

```
plt.scatter(X_test, y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue') # This doesn't have to be changed as the regression line is unique
plt.title('Salary vs Experience (Predicted set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (R)')
plt.show()
```
